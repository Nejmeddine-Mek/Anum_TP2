```python
def OGD(A,b,epsilon):
    n = A.shape[0]
    X = np.zeros(n).reshape(n,1)
    r = b - A @ X

    while(np.linalg.norm(r) >= epsilon):
        temp = r.flatten()
        a = np.dot(temp,temp) / np.dot((A @ r).flatten(),temp)
        X += a * r
        r = b - A @ X
    
    return X
```

```python
A = np.array([
    [4, 1, 1],
   [1, 3, -1],
    [1, -1, 2]
], dtype=float)

b = np.array([6, 3, 2], dtype=float)
b = b.reshape(b.shape[0],1)

res = OGD(A,b,0.00001)
print(res)
```
