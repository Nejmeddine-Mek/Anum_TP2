```python
import numpy as np
```

```python
def GaussSeidel(A,b,epsilon):
    n = A.shape[0]
    D = np.diag(np.diag(A))
    U = -np.triu(A,1)
    L = -np.tril(A,-1)
    Lgs = np.linalg.inv(D - L)
    Tgs = Lgs @ U
    radius = np.linalg.eigvals(Tgs)
    if(np.max(abs(radius)) < 1):
        X_cur = np.zeros(n).reshape(n,1)
        r = b - A @ X_cur
        while(np.linalg.norm(r) >= epsilon):
            X_cur = Lgs @ (U @ X_cur + b)
            r = b - A @ X_cur
        return X_cur
    else:
        return 0
```

```python
A = np.array([[5,1,0],[2,5,1],[1,1,6]])
GaussSeidel(A,[[6],[-4],[9]],10 ** (-4))
```
